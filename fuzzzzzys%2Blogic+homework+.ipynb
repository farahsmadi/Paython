```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
#function for trapozidal
def trapmemberf(x,a,w,r,e):
    if (x <=a ):
        m=0
    elif (x> a and x<w):
        m=(x-a)/(w-a)
       
    elif (x>= w and x<=r):
        m=1
    elif (x>r and x<e):
        m=(e-x)/(e-r)
    elif(x>=e):
        m=0
        
    return m
```

```python
#function for triangle
def tramfun(x,y,b,q):
    if(x < y):
        z=0
    elif(x>=y and x<=b):
        z=(x-y)/(b-y)
    elif(x>=b and x<=q):
        z=(q-x)/(q-b)
    elif(x>q):
        z=0
    return z
```

```python
x=np.arange(150,211)
```

```python
short =[trapmemberf(i,150,155,163,170) for i in x]
```

```python
avg=[tramfun(i,165,175,185) for i in x]
```

```python
tall=[trapmemberf(i,180,190,210,240) for i in x]
```

```python
#draw our figure
plt.figure(figsize=(12, 6))
_=plt.plot(x,short,color='blue')
_=plt.plot(x,avg,color='red',)
_=plt.plot(x,tall,color='blue')
plt.ylabel('Fuzzy membership')
plt.xlabel('hight')
_ = plt.legend(('short', 'avg', 'tall'), loc='center right', prop={'size': 20})

plt.show()

```
